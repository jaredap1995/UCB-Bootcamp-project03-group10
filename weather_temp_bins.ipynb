```python
import pandas as pd
from pathlib import Path
```

```python
joined_tables_rawdata = Path("back_end/joined_tables.csv")
```

```python
joined_tables = pd.read_csv(joined_tables_rawdata)
```

```python
joined_tables.head()
```

```python
dtype_before = type(joined_tables["temp_real_f"])
  
# converting to list
temp_list = joined_tables["temp_real_f"].tolist()
  
# storing dtype after operation
dtype_after = type("temp_real_f")
  
temp_list
```

```python
joined_tables_df = pd.DataFrame(joined_tables)
joined_tables_df.head()
```

```python
sorted_version_desc = joined_tables_df.sort_values(by='temp_real_f', ascending=False)
sorted_version_desc.head()
```

```python
sorted_version_asc = joined_tables_df.sort_values(by='temp_real_f', ascending=True)
sorted_version_asc.head()
```

```python
bins = [0, 30, 50, 70, 90, 110]

# Create the names for the five bins
group_names = ["0-30F", "31-50F", "51-70F", "71-90F", "91-110F"]
```

```python
bins_data = joined_tables_df["temp_real_f"] = pd.cut(joined_tables_df["temp_real_f"], bins, labels=group_names, include_lowest=True)
bins_data
```

```python
group_df = joined_tables_df.groupby("temp_real_f")
group_df.sum()
```

```python
df_sum = group_df.sum()
df_sum
```

```python
df_sum.drop(columns=['id', 'id-2', 'id-3','id-4'], axis=1, inplace=True)
#df_sum.drop(columns=['temp_feel_f','precipitation_inch','weathercode','wind_speed'], axis=1, inplace=True)
df_sum
```

```python
df_sum.dtypes
```

```python
convert_dict = {'north_south_sum_broadway': int,
            'north_bike_broadway': int,
            'south_bike_broadway': int,
            'total_ped_and_bike_burke': int,
            'pedestrian_south_burke': int,
            'pedestrian_north_burke': int,
            'bike_north_burke': int,
            'bike_south_burke': int,
            'east_west_sum_fremont': int,
            'east_bike_fremont': int,
            'west_bike_fremont': int,                
            }
df_sum = df_sum.astype(convert_dict)
print(df_sum.dtypes)
```

```python
# Save the resulting DataFrame to a CSV file
df_sum.to_csv("back_end/weather_temp_bins.csv", encoding="utf-8", index=True, header=True)
```
